EXPLORATORY DATA ANALYSIS
WHAT IS OVERALL SALES TREND?

```python
# Min date in the data & Max date in the data
print("Minimum Date:",df['Order Date'].min())
print("Maximum Date:",df['Order Date'].max())
```

Approach
1.	Generate a new column with Year (YYYY)
2.	Group all the Sales of the same Year and combine it with the above New Column
3.	Plot the new dataframe with X axis as Year and Y axis as Combined Sales (Total Cost).
 

```python
# 1. Generate a new column of Month year 
df['Month Year']=df['Order Date'].apply(lambda x:x.strftime('%Y-%m')) # changing format of the date
df
```

```python
#Drop Year Column
df.drop(columns =['Year'])
```

```python
# 2. New Data frame with Year and combined sales 
df_trend = df.groupby('Month Year').sum()['Total Cost'].reset_index() # The reset_index() method allows you reset the index back to the default 0, 1, 2 etc indexes.
df_trend 
```

```python
# 3. setting figure sizeand colour
plt.figure(figsize=(20,9))
plt.plot(df_trend['Month Year'],df_trend['Total Cost'], color = 'red')
plt.xticks(rotation = 'vertical', size = 10)
plt.show()

```

# Inferences
1. In all years of sales are not that much high expect year 2013 Feb Month.
2. In all the year sales, there is a specific drop in sales in May to June as well as in october.

# WHICH ARE TOP 5 ITEM TYPE BY TOTAL COST?


Approach
1. We need two columns, Iteam Type and the Total Cost.
2. We will get the combined Total Cost for that project using groupby
3. Arrange it in descending order and print out top 5 objects.


```python
# create a new data frame
product_sales=pd.DataFrame(df.groupby('Item Type').sum()['Total Cost'])
product_sales
```

```python
# Sorting the sales in descending order
product_sales = product_sales.sort_values('Total Cost',ascending=False)
```

```python
# Displaying only top 5 rows
product_sales[:5]
```

# Inferences
These are the top 5 Items with their corresponding sales(Total Cost), which we can furthur use to implement our marketing strategies.

# WHICH ARE MOST SELLING Product?
Approach
1. We need a table which contain Item Type and the combined Unit Sold.
2. Group that into a dataframe.
3. Arrange them in descending order and print only first 3.

```python
# Generating required Data table

product_most_sold = pd.DataFrame(df.groupby(['Item Type']).sum()['Units Sold'].reset_index())

product_most_sold
```

```python
# Sorting the data of Units Sold column in descending order
product_most_sold = product_most_sold.sort_values('Units Sold', ascending = False)
```

```python
# Displaying Top 3 most selling products

product_most_sold[:3]
```
